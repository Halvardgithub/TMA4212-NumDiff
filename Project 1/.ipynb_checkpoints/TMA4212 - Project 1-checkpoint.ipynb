# Project 1



In this project we will look into modelling the heat distribution in anisotropic materials. By Fourier's law, we are given the following model:

$$
-\nabla \cdot (\kappa \nabla u) =f \qquad \text{ in } \Omega
$$

We can reformulate this problem as 

$$
-(a+1)\partial_x^2u - 2r\partial_x\partial_yu - r^2 \partial_y^2u=-a \partial_x^2u - (\textbf{d}_2\cdot\nabla)^2 u = f
$$

where $\textbf{d}_2=(1,r)^\top$

```python
#importing libraries
import numpy as np
from scipy.sparse import diags, bmat
from scipy.sparse.linalg import spsolve
from matplotlib import cm
```

```python
def plot_solution(x, y, U, txt='Solution'):
    fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
    X, Y = np.meshgrid(x,y)
    ax.plot_surface(X, Y, U, cmap=cm.coolwarm)
    #ax.view_init(azim=0)              # Rotate the figure
    plt.xlabel('x')
    plt.ylabel('y')
    plt.tight_layout()
    plt.title(txt)
```

In this task, we are working with a PDE of the form
$$
-\mathcal{L} u = -k u_{xx}-2bu_{xy}-cu_{yy}-du_x-eu_y-lu = f \qquad \in \Omega
$$

we need to discretize this, i.e. find coefficients $\alpha_{PQ_i}$ such that

$$
\begin{aligned}
-\mathcal{L} u_p &= -\mathcal{L}_h u_p + \tau_p\\
-\mathcal{L} U_p &= \alpha_{pp}U_p - \sum_{i=1}^s \alpha_{PQ_i} U_{Q_i}
\end{aligned}
$$

from the lectures, we know that we can determine these coefficients by following the conditions underneath (following from a Taylor expansion)

$$
\begin{aligned}
h^0u_p: \qquad &\alpha_{pp}-\sum_{i=1}^s\alpha_{PQ_i}=-l_p\\
h^1u_x: \qquad &-h\sum_{i=1}^s\alpha_{PQ_i} \xi_i=-d_p\\
h^1u_y: \qquad &-h\sum_{i=1}^s\alpha_{PQ_i} \eta_i=-e_p\\
h^2u_{xx}: \qquad &-\frac{h^2}{2}\sum_{i=1}^s\alpha_{PQ_i} \xi_i^2=-k_p\\
h^2u_{xy}: \qquad &-h^2\sum_{i=1}^s\alpha_{PQ_i} \xi_i \eta_i=-2b_p\\
h^2u_{yy}: \qquad &-\frac{h^2}{2}\sum_{i=1}^s\alpha_{PQ_i} \eta_i^2=-c_p\\
\end{aligned}
$$

we have chosen to split our differential equation into two different cases, by looking at

$$
-\mathcal{L}_1 u = - a \partial_x^2 u \qquad -\mathcal{L}_2 u = -(\textbf{d}_2\cdot\nabla)^2 u
$$

for $\mathcal{L}_1$, we use the usual central difference scheme, which is known as

$$
-\mathcal{L}_1 u_m^n =  -\frac{a}{h^2} \left(u_{m+1}^n - 2u_m^n + u_{m-1}^n \right) + O(h^2)
$$

for $\mathcal{L}_2$, we also use a central difference scheme, with points along the directional derivative:

$$
-\mathcal{L}_{2_h} U_m^n = \alpha_0 U_m^n - \alpha_1 U_{m-1}^{n-1} - \alpha_2 U_{m+1}^{n+1} 
$$

For our problem, we have $k=1, b=2, c=4, l=e=d=0$.
With $r=2$ we get a step size of $\sqrt{5}h$

we now have the following conditions:

$$
\begin{aligned}
\alpha_0-\alpha_1-\alpha_2 = 0\\
-\sqrt{5}h \left(\alpha_1\frac{1}{\sqrt{5}} - \alpha_2\frac{1}{\sqrt{5}}\right) = 0\\
-\sqrt{5}h \left(\alpha_1\frac{2}{\sqrt{5}} - \alpha_2\frac{2}{\sqrt{5}}\right) = 0\\
-\frac{5h^2}{2} \left(\alpha_1\frac{1}{5} + \alpha_2\frac{1}{5}\right) = -1\\
- 5h^2 \left(\alpha_1\frac{2}{5} + \alpha_2\frac{2}{5}\right) = -4\\
-\frac{5h^2}{2} \left(\alpha_1\frac{4}{5} + \alpha_2\frac{4}{5}\right) = -4\\
\end{aligned}
$$

this gives $\alpha_0 = \frac{2}{h^2}, \alpha_1=\alpha_2=\frac{1}{h^2}$, so we have

$$
-\mathcal{L}_{2_h} U_m^n =  -{h^2} \left(U_{m+1}^{n+1} - 2U_m^n + U_{m-1}^{n-1} \right)
$$

in total, we have the scheme

$$
-\mathcal{L}_h U_m^n = -a \frac{1}{h^2} (U_{m+1}^n - 2U_m^n + U_{m-1}^n )-\frac{1}{h^2}(U_{m+1}^{n+1} - 2U_m^n + U_{m-1}^{n-1})
$$

To write this in matrix form, we first need to consider an ordering of the points. We have chosen to use a natural ordering, to translate the matrix $U$ into a vector $\textbf{U}$ consisting of the $(M-1)^2$ interior points. We then have

$$
A\textbf{U} = h^2 \textbf{f} + \textbf{b}
$$

with

$$
\begin{aligned}
B &= \text{tridiag}\left\{-a,2(a+1),-a\right\}\\
C &= \text{tridiag}\left\{0, 0 ,-1\right\}\\
\\
A &= \text{tridiag} \left\{C^\top, B, C\right\}
\end{aligned}
$$

In order to determine $\textbf{b}$, we have started with considering a vector $\textbf{g}$ which consists of the boundary values , again in a natural ordering. Our idea was then to find a matrix $B\in \mathbb{R}^{(M-1)^2\times 4M}$ that transforms our vector $\textbf{g}$ into $\textbf{b}$. Here we'll have $(M-1)^2$ rows to represent each of our interior points, and $4M$ since that is how many boundary values (BV) we will have. In order to determine $B$ we have considered the following model/illustration:



If we first consider the row right above the lower boundary, we see that for the first corner here will need the first BV, and that we can ignore the following $M$ (here equal to 5) BVs, before we again need the next BV. So we can express the row in $B$ corresponding to this corner as
$$
\left[s \quad \textbf{0} \in\mathbb{R}^M \quad H \quad \textbf{0}\in \mathbb{R}^{3M-2}\right]
$$
Here the values $s$ and $H$ represent the scaling these values get from our difference scheme, and are $\frac{1}{5}$ and $a$. For the horizontal non-corners we first note that we will have $M-3$ of them, meaning that they will make up this many rows of $B$ as well. However, since each of them only depend on one boundary value, we can represent each row number i as 
$$
\left[\textbf{0}\in\mathbb{R}^i \quad s \quad \textbf{0}\in\mathbb{R}^{4M-(i+1)}\right] \quad i\in [1, 3M-1]
$$
For the last corner in this row, the situation is a bit more tricky. Here we will first have to skip the $M-2$ first BVs, use the next BV, and then skip three again until we use the one to the right for our point, before we again skip one, and use the one to the top-right, meaning that the row will be
$$
\left[ \textbf{0}\in\mathbb{R}^{M-2} \quad s \quad \textbf{0}\in\mathbb{R}^3 \quad H \quad 0 \quad s \quad \textbf{0}\in \mathbb{R}^{3M-5}  \right]
$$

Combining these rows into a matrix row-wise gives a matrix that when multiplied to $\textbf{g}$ gives the scaled boundary values according to our scheme for all the points on the row we are considering now. 

Furthermore, due to the symmetry of our problem, we actually have that for the low right below the upper boundary, is the "flipped" version of the matrix presented above, where "flipping" denotes the following operation, where the rightmost matrix is the "flipped" matrix.

$$
\begin{bmatrix}
a && b && c \\
d && e && f \\
g && h && i 
\end{bmatrix}
\to
\begin{bmatrix}
c && b && a \\
f && e && d \\
i && h && g
\end{bmatrix}
\to
\begin{bmatrix}
i && b && g \\
f && e && d \\
c && h && a
\end{bmatrix}
$$

So until now we have established that our $B$ matrix is a block matrix, consisting of  the matrix presented above at its top, and the "flipped" matrix at its bottom. So now we have to decide which matrices we have to fill into the middle of $B$. In order to do this, we will define one matrix for each row/grid line consisting of orange and purple points, meaning that we will have $M-3$ of these matrices. Since each of our grid lines consists of two purple points and $M-3$ orange points we will have that the $M-3$ interior rows of these matrices will only be the zero vector, but that the first and last row will be different from zero. 

To implement this, we made a general formula for the i-th row after the inner gridline, so our second row of inner points will have i=1. Since we only get boundary contributions from the edges, we only need to calculate four of these indecies per row-matrix. The generel formula for the contrebutions on the left is$indexDiag_l = M + 2i$ and $ indexHor_l = M + 2(i+1)$ while for the contributions on the right side we get $ indexDiag_r = M + 2(i + 2) + 1 $ and $ indexHor_r = M + 2(i+1) + 1$. The left indecies are plugged in on the first row of our matrix while the right contributions are plugged in on the last row.

```python
def tridiag_sparse(a,b,c,N):
    bidiag_l = np.full(N-1, a)
    diag = np.full(N, b)
    bidiag_u = np.full(N-1, c)
    
    A = diags([bidiag_l, diag, bidiag_u],[-1,0,1])
    return A
```

```python
def block_tridiag_sparse(A, B, C):
    N = B.shape[0]
    A = bmat([[B if i == j else A if i-j==-1 else C if i-j==1 else None for i in range(N)] for j in range(N)], format='csc')
    #A = bmat([[B if i == j else A if i-j==-1 else C if i-j==1 else None for i in range(N)] for j in range(N)], format='csc')
    return A
```

```python
def g(x,y):
    lower = x*2
    upper = x*4
    left = y*0
    right = y**2
    
    boundary = {"left": left, "right": right, "lower":lower, "upper": upper}
    return boundary
```

```python
M=10
x=np.linspace(0,1,M+1)
y=np.linspace(0,2,M+1)
print(g(x, y))
```

```python

def upper_mat(M, s, H):
    row1 =  np.concatenate(([s], np.zeros(M), [H], np.zeros(3*M-2)))
    inner_mat = np.c_[np.zeros(M-3), np.eye(M-3)*s, np.zeros((M-3, 3*M+2))]
    row2 = np.concatenate((np.zeros(M-2), [s], np.zeros(3), [H], [0], [s], np.zeros(3*M-5)))
    #print(row1, "\n")
    #print(inner_mat, "\n")
    #print(row2, "\n")
    return np.r_[[row1], inner_mat, [row2]]
    
#print(upper_mat(1000, 1, 2))

def I(M, i, s, H):
    assert i<(M-2), "i too low"
    I = np.zeros((M-1, 4*M))
    index_s = M + 2*i - 1
    index_H = index_s + 3
    I[0, index_s] = s
    I[0, index_s + 2] = H
    I[-1, index_H] = H
    I[-1, index_H + 2] = s
    return I
    
def build_B(M,s, H):
    mat1 = upper_mat(M, s, H)
    #print(mat1)
    mat2 = np.flip(np.flip(mat1, 0), 1)
    #print(mat2)
    B = mat1
    
    for i in range(1, M-2):
        B = np.r_[B, I(M, i, s, H)]
    return np.r_[B, mat2]

plt.imshow(build_B(6, 1, 2))
```

```python
def WIP(g, f, M, a):
    h = 1/M
    k = 2*h
    
    print(f"h = {h:.4}, k = {k:.4}")
    
    x = np.linspace(0,1,M+1)
    y = np.linspace(0,2,M+1)
    g = g(x,y)
    
    U = np.zeros((M+1, M+1))
    U[0,:] = g["lower"]
    U[-1,:] = g["upper"]
    U[:,0] = g["left"]
    U[:,-1] = g["right"]
    
    #X, Y = np.meshgrid(x, y)
    g_vec = np.zeros(4*M)
    g_vec[:M+1] = g["lower"]
    g_vec[-(M+1):] = g["upper"]
    
    j=1
    for i in range(1,2*(M-1),2):
        g_vec[M+i] = g["left"][j]
        g_vec[M+i+1] = g["right"][j]
        j+=1
    #print(g_vec)
    #diags of B
    B = tridiag_sparse(-a, 2*(a+1), -a, M-1)
    C = diags(np.full(M-2,-1),1)
    
    A = block_tridiag_sparse(C.T, B, C)
    
    X, Y = np.meshgrid(x[1:-1], y[1:-1])
    #print(f(X, y, a))
    f_vec = np.ravel(f(X,Y, a))
    #f_vec = np.ravel(f(x, y, a))
    b_vec = build_B(M, 1, a)@g_vec
    
    U[1:-1,1:-1] = np.reshape(spsolve(A, b_vec+f_vec*h**2), (M-1,M-1))
    
    return x, y, U
```

```python
M=100
import matplotlib.pyplot as plt

x = np.linspace(0,1,M+1)
y = np.linspace(0,2,M+1)

X, Y = np.meshgrid(x, y)

print(X)
print(Y)

def f(x,y, a):
   # U = -np.cos(X)*np.sin(Y)*(a+1)-4*np.cos(Y)*np.sin(X)-4*np.cos(X)*np.sin(Y)
    #U = -np.pi**2*np.sin(np.pi*X)*np.sin(np.pi*Y)*(a+1)+4*np.pi**2*np.cos(np.pi*X)*np.cos(np.pi*X)-4*np.pi**2*np.sin(np.pi*X)*np.sin(np.pi*Y)
    #U=np.sin(np.pi * x) * np.sin(np.pi * y) - np.pi**2 * np.sin(np.pi * x) * np.sin(np.pi * y) * (a + 1) + (2 * 2) * np.pi**(2) * np.cos(np.pi * x) * np.cos(np.pi * y)
    a1 = np.pi**2*np.sin(np.pi*x)*np.sin(np.pi*y)*(a+1)
    a2 = -4*np.pi**2*np.cos(np.pi*x)*np.cos(np.pi*y)
    a3 = 4*np.pi**2*np.sin(np.pi*x)*np.sin(np.pi*y)
    return (a1+a2+a3)

def u_exact(x, y):
    return np.sin(x*np.pi)*np.sin(y*np.pi)
```

```python
x, y, U = WIP(1, f, 1000, 1)

X, Y = np.meshgrid(x[1:-1], y[1:-1])
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, U, cmap=cm.coolwarm)
#ax.view_init(azim=0)
```

```python
def convergence(u_exact, g, f, M=100, P=4): 
    Hconv, Econv = np.zeros(P), np.zeros(P)
    for p in range(P):
        x, y, U = WIP(g, f, M, 1)
        #x = x[1:-1]
        #y = y[1:-1]
        X, Y = np.meshgrid(x,y)
        Eh = u_exact(X,Y)-U
        Hconv[p] = (x[1]-x[0])
        M = 2*M
        Econv[p] = np.linalg.norm(Eh, ord = np.inf)
        np.max(np.abs(Eh))
    order = np.polyfit(np.log(Hconv),np.log(Econv),1)[0]
    return Hconv, Econv, order
```

```python
H, E, p = convergence(u_exact, 1, f)
plt.loglog(H,E,'o-', label='p={:.2f}'.format(p))
plt.grid('on')
plt.xlabel('h')
plt.ylabel('error')
plt.legend()
plt.show()
```
